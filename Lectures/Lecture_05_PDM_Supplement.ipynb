>## Deriving the Posterior for the Normal–Normal Model

We observe data  
$$
x_1, x_2, \dots, x_n \sim \mathcal{N}(\theta, \sigma^2),
$$
where $\sigma^2$ is known.  
We place a normal **prior** on the unknown mean:
$$
\theta \sim \mathcal{N}(\mu_0, \tau^2).
$$

---

### **1. Likelihood**

### Likelihood

The likelihood function for $\theta$ based on the observed data is

$$
L(\theta \mid x) = p(x \mid \theta)
\propto
\exp\!\left(
-\frac{1}{2\sigma^2}
\sum_{i=1}^n (x_i - \theta)^2
\right).
$$

Although we write $p(x \mid \theta)$, in likelihood theory the data $x$ are
treated as fixed and the expression is viewed as a *function of the
parameter* $\theta$.

Expanding the sum:
$$
\sum_{i=1}^n (x_i - \theta)^2
= \sum x_i^2 \;-\; 2\theta\sum x_i \;+\; n\theta^2.
$$

Ignoring terms that do not depend on $\theta$, the likelihood is proportional to
$$
\exp\!\left(
-\frac{1}{2\sigma^2}
\left[n\theta^2 - 2\theta\sum x_i\right]
\right).
$$

---

### **2. Prior**

The normal prior can be written as
$$
p(\theta)
\propto
\exp\!\left(
-\frac{1}{2\tau^2}(\theta - \mu_0)^2
\right)
=
\exp\!\left(
-\frac{1}{2\tau^2}(\theta^2 - 2\mu_0\theta)
\right),
$$
again dropping constants.

---

### **3. Posterior is proportional to likelihood × prior**

Multiply exponents:
$$
\log p(\theta \mid x)
= -\frac{1}{2}
\left(
\frac{n}{\sigma^2} + \frac{1}{\tau^2}
\right)\theta^2
\;+\;
\left(
\frac{\sum x_i}{\sigma^2} + \frac{\mu_0}{\tau^2}
\right)\theta
\;+\; \text{const}.
$$

Define:
$$
A = \frac{n}{\sigma^2} + \frac{1}{\tau^2},
\qquad
B = \frac{\sum x_i}{\sigma^2} + \frac{\mu_0}{\tau^2}.
$$

Then:
$$
\log p(\theta \mid x)
= -\frac{A}{2}\theta^2 + B\theta + \text{const}.
$$

---

### **4. Completing the square**

We rewrite the quadratic form as:
$$
-\frac{A}{2}\theta^2 + B\theta
=
-\frac{A}{2}\left(\theta - \frac{B}{A}\right)^2 + \text{const}.
$$

Thus the posterior distribution (prob parameters given data) is normal:
$$
\theta \mid x \sim \mathcal{N}(\mu_n, \tau_n^2)
$$
with
$$
\mu_n = \frac{B}{A},
\qquad
\tau_n^2 = \frac{1}{A}.
$$

Substituting back:
$$
\tau_n^2 = 
\left(
\frac{n}{\sigma^2} + \frac{1}{\tau^2}
\right)^{-1},
$$

$$
\mu_n
= \tau_n^2\left(
\frac{n\bar{x}}{\sigma^2} + \frac{\mu_0}{\tau^2}
\right),
\quad
\text{where } \bar{x} = \frac{1}{n}\sum x_i.
$$

---

### **5. Interpretation**

The posterior variance is
$$
\tau_n^2 = 
\left(
\underbrace{\frac{n}{\sigma^2}}_{\text{data precision}}
+
\underbrace{\frac{1}{\tau^2}}_{\text{prior precision}}
\right)^{-1},
$$
so the total **precision** (inverse variance) is the sum of data and prior precisions.

The posterior mean is a precision-weighted average:
$$
\mu_n
= 
\left[
\frac{n/\sigma^2}{n/\sigma^2 + 1/\tau^2}
\right]\bar{x}
\;+\;
\left[
\frac{1/\tau^2}{n/\sigma^2 + 1/\tau^2}
\right]\mu_0.
$$

For large $n$, the data dominate and $\mu_n \to \bar{x}$.  
For small $n$, the prior strongly influences $\mu_n$.

--

When you plug in the numbers you will get a posterior distribution with $\tau_n = 6.7$. The mean is then given by $\mu_n = 147.6$. These are the numbers shown in the first figure.
