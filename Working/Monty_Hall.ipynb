```python
import random as rand
```

```python
games = 100000
noswitch_win = 0
switch_win = 0
for i in range(0,games):
    doors = ['door', 'door', 'door']
    doors[rand.randint(0,2)] = 'car'


    if doors[1] == 'car':
        doors.pop(2)
    elif doors[2] == 'car':
        doors.pop(1)
    else:
        doors.pop(rand.randint(1,2))


    if doors[0] == 'car':
        noswitch_win += 1
    else:
        switch_win += 1

print(noswitch_win/games)
print(switch_win/games)
```
